### Answer here:
Since the vocabulary and count of document is large, it is not useful to plot out every single document. More useful would be to view a sample of the document at a time, and apply a threshold for frequencies (e.g. only view words with frequency > 10).  

Below I define a function that get only words with count > 0 from the designated sample of the matrix (the range of smaples is assigned with `doc_incides_list`)  

```python
#X_counts: the word count vector for every doc (num_docs*total_num_wrds)

def get_X_counts_with_val(X_counts, count_vect, doc_indices_list = []):
    """
    idea: get only words that have values
    """
    if not doc_indices_list:
        docs = X_counts[:]
    else:
        docs = X_counts[doc_indices_list]
    res_X_counts = np.zeros(len(count_vect), dtype=int)
    for doc in docs:
        doc_np = doc.toarray()
        if doc_np.shape == res_X_counts.shape:
            res_X_counts = (doc_np | res_X_counts) # OR-ing gives max value of the 2 arrays
        else:
            res_X_counts = (doc_np[0] | res_X_counts)
    return res_X_counts
```

Then I apply a mask to extract count values that meet a certain frequency count threshold (e.g. frequency > 10)

```python
feat_name_arr = np.array(count_vect.get_feature_names())
doc_ids = list(range(2,10))
threshold_mask = res_cnt > 10 #(res_cnt > 4) & (res_cnt < 8)

res_cnt = get_X_counts_with_val(X_counts, feat_name_arr, doc_ids)
words = feat_name_arr[threshold_mask]
res_docs = np.asarray([doc.toarray()[0][threshold_mask] for doc in X_counts[doc_ids]])
```

## 1, 2. Load & Prepare Data

```python
def load_sep_data(filepath): 
    """
    load data and split sentences by linebreak.
    """
    ret_dict = dict()
    with open(filepath) as f:
        sents = []
        scores = []
        line = f.readline()
        while line:
            sent, score = line.split("\t")
            sents.append(sent)
            scores.append(int(score))
            line = f.readline()
        ret_dict['sentences'] = sents
        ret_dict['scores'] = scores
    return ret_dict
```

```python
orig_data = dict()
orig_data['amazon'] = load_sep_data('./senti_analysis_data/amazon_cells_labelled.txt')
orig_data['imdb'] = load_sep_data('./senti_analysis_data/imdb_labelled.txt')
orig_data['yelp'] = load_sep_data('./senti_analysis_data/yelp_labelled.txt')
```

```python
#X_counts: the word count vector for every doc (num_docs*total_num_wrds)

def get_X_counts_with_val(X_counts, count_vect, doc_indices_list = []):
    """
    idea: get only words that have values
    """
    if not doc_indices_list:
        docs = X_counts[:]
    else:
        docs = X_counts[doc_indices_list]
    res_X_counts = np.zeros(len(count_vect), dtype=int)
    for doc in docs:
        if type(doc).__module__ != np.__name__:
            doc = doc.toarray()
        if doc.shape == res_X_counts.shape:
            res_X_counts = (doc | res_X_counts) # OR-ing gives max value of the 2 arrays
        else:
            res_X_counts = (doc[0] | res_X_counts)
    return res_X_counts
```

```python
def get_max_tfidf_of_all_docs(X_tfidf):
    
    X_tfidf = X_tfidf.toarray()
    res_mask = np.zeros(len(X_tfidf[0]))
    
    for doc in X_tfidf:
        for i, feat_val in enumerate(doc):
            # getting max across all docs is prob not a good idea...
            # esp when we want ot eliminate stopwords
            # => only get max if its val isn't 0
            if feat_val:
                res_mask[i] = max(feat_val, res_mask[i]) 
            
    return res_mask
```

```python
tfidf_mask_nostop = get_max_tfidf_of_all_docs(X_tfidf_nostop)
```

- **Bernoulli**: a binary distribution useful when a feature can be present or absent.   
- **Multinomial**: a discrete distribution used whenever a feature must be represented by a whole number (for example, in natural language processing, it can be the frequency of a term) --> **Term freq**  
- **Gaussian**: a continuous distribution characterized by its mean and variance. --> **TF-IDF**

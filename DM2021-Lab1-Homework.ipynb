## 1, 2. Load & Prepare Data

```python
def load_sep_data(filepath):
    ret_dict = dict()
    with open(filepath) as f:
        sents = []
        scores = []
        line = f.readline()
        while line:
            sent, score = line.split("\t")
            sents.append(sent)
            scores.append(int(score))
            line = f.readline()
        ret_dict['sentences'] = sents
        ret_dict['scores'] = scores
    return ret_dict
```

```python
orig_data = dict()
orig_data['amazon'] = load_sep_data('./senti_analysis_data/amazon_cells_labelled.txt')
orig_data['imdb'] = load_sep_data('./senti_analysis_data/imdb_labelled.txt')
orig_data['yelp'] = load_sep_data('./senti_analysis_data/yelp_labelled.txt')
```

```python
#X_counts: the word count vector for every doc (num_docs*total_num_wrds)

def get_X_counts_with_val(X_counts, count_vect, doc_indices_list = []):
    """
    idea: get only words that have values
    """
    if not doc_indices_list:
        docs = X_counts[:]
    else:
        docs = X_counts[doc_indices_list]
    res_X_counts = np.zeros(len(count_vect), dtype=int)
    for doc in docs:
        doc_np = doc.toarray()
        if doc_np.shape == res_X_counts.shape:
            res_X_counts = (doc_np | res_X_counts) # OR-ing gives max value of the 2 arrays
        else:
            res_X_counts = (doc_np[0] | res_X_counts)
    return res_X_counts
```

```python
def get_max_tfidf_of_all_docs(X_tfidf):
    
    X_tfidf = X_tfidf.toarray()
    res_mask = np.zeros(len(X_tfidf[0]))
    
    for doc in X_tfidf:
        for i, feat_val in enumerate(doc):
            # getting max across all docs is prob not a good idea...
            # esp when we want ot eliminate stopwords
            # => only get max if its val isn't 0
            if feat_val:
                res_mask[i] = max(feat_val, res_mask[i]) 
            
    return res_mask
```
